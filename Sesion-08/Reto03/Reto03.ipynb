Crea un descriptor de color utilizando tres histogramas. Pruebalo con tres im√°genes de manzanas y tres de bananas. Los descriptores de manzanas deben ser similares entre si, al igual que los descriptores de bananas.

Para ver si son similares, grafica con plt.bar() y con plt.ylim(0,500). 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def abrirImagen(ruta, mostrar = False):
    imageOpenCV = cv2.imread(ruta, cv2.IMREAD_COLOR)
    imagen = cv2.cvtColor(imageOpenCV, cv2.COLOR_BGR2RGB)
    if(mostrar):
        plt.figure(figsize=(7,7))
        plt.imshow(imagen)
        plt.show()
    return imagen
```

```python
imgManzana1 = abrirImagen("manzana.jpg", mostrar = True)
imgManzana2 = abrirImagen("manzana2.jpg", mostrar = True)
imgManzana3 = abrirImagen("manzana3.jpg", mostrar = True)
```

```python
imgBanana1 = abrirImagen("banana.jpg", mostrar = True)
imgBanana2 = abrirImagen("banana2.jpg", mostrar = True)
imgBanana3 = abrirImagen("banana3.jpg", mostrar = True)
```

```python
def crearHistograma(imagen):
    histograma = np.zeros((256))
    imgEnArray = np.ravel(imagen)
    for counter in range(0, len(imgEnArray)):
        histograma[int(imgEnArray[counter])]+=1
    return histograma
```

```python
def crearDescriptor(imagen):
    histogramaRojo = crearHistograma(imagen[:,:,0])
    histogramaVerde = crearHistograma(imagen[:,:,1])
    histogramaAzul = crearHistograma(imagen[:,:,2])
    descriptor = np.concatenate([histogramaRojo, histogramaVerde, histogramaAzul])
    print(descriptor.shape)
    return descriptor
```

```python
descriptorManzana1 = crearDescriptor(imgManzana1)
plt.ylim(0,500)
plt.bar(np.arange(len(descriptorManzana1)),descriptorManzana1, color='red')
plt.show()

descriptorManzana2 = crearDescriptor(imgManzana2)
plt.ylim(0,500)
plt.bar(np.arange(len(descriptorManzana2)),descriptorManzana2, color='red')
plt.show()

descriptorManzana3 = crearDescriptor(imgManzana3)
plt.ylim(0,500)
plt.bar(np.arange(len(descriptorManzana3)),descriptorManzana3, color='red')
plt.show()
```

```python
descriptorBanana1 = crearDescriptor(imgBanana1)
plt.ylim(0,500)
plt.bar(np.arange(len(descriptorManzana1)),descriptorBanana1, color='orange')
plt.show()

descriptorBanana2 = crearDescriptor(imgBanana2)
plt.ylim(0,500)
plt.bar(np.arange(len(descriptorBanana2)),descriptorBanana2, color='orange')
plt.show()

descriptorBanana3 = crearDescriptor(imgBanana3)
plt.ylim(0,500)
plt.bar(np.arange(len(descriptorBanana3)),descriptorBanana3, color='orange')
plt.show()
```
